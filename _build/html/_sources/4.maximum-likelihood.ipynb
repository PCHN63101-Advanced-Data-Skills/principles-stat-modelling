# Maximum Likelihood Estimation
In the previous section, we saw how conceptualising our measurements as a random variable led to a model of the *data-generating processing*. This model centres around some probability distribution with a finite number of parameters. These parameters relate to both the *expected value* and the *variance* of the distribution, which in turn tell us everything that we need to know about the variable in question. This can be thought of as a *simplification* process, where the information within the $n$ datapoints is summarised by the $k$ parameters of the distribution. As every datapoint can be conceptualised as a combination of the expected value and random variation, these parameters can be used to explain every value we measure. As such, the parameters of the proposed distrubution are *the most important element of our statistical model*.


## Estimating Parameters
Given the importance of the parameters, we of course what to know what their values are. Unfortunately, we do not often have the whole population at our disposal. If we did, we could simply calculate the parameter values and be done. More typically, we only have a *sample* of values from the distribution and therefore must use the information within the same to *estimate* the parameter values.

``````r
set.seed(666)

y          <- rnorm(n=100, mean=50, sd=1)
mu.ests    <- seq(40,60,length.out=1000) # 1000 possibilities for mu
likelihood <- rep(0,length(mu.ests))

for (i in 1:length(mu.ests)){
    likelihood[i] <- prod(dnorm(y, mean=mu.ests[i], sd=1))
}

plot(mu.ests,likelihood, type="l")

max.idx <- which.max(likelihood)
mu.ests[max.idx]
``````
