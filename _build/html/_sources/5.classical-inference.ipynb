## Manipulating Random Variables
... It is important to recognise that when a random variable is manipulated in some way, the fact that it is random does not change. However, the specific distribution *will* change. ... A lot of effort within classical statistics has been dedicated to manipulating random variables and then trying to figure out what distribution the manipulated value has.

As an example, it is a known result in mathematical statistics that if you square a random variable that has a standard normal distribution, the resultant random variable has a $\chi^{2}$ distribution. So, formally, we have

$$
y \sim \mathcal{N}\left(0,1\right)
$$

and

$$
y^{2} \sim \chi^{2}\left(1\right).
$$

The importance of this is to note two things. One is that transforming or manipulating a random variable does *not* remove the randomness, it just changes it. As such, any equation that contains a random variable will result in another random variable. This will be important going forward. The second is that much of classical statistics is constrained by requiring the distribution of the manipulated random variable to be known. As such, many of the methods and results that we used are based on these distributions being *tractable*. If no one has worked out what the distribution is, the manipulation cannot be used. This restriction can explain a lot of the choices made within the context of statistical modelling. For instance, the normal distribution is used so frequently because there are a multitude of results that are mathematically convenient, allowing these forms of distributions to be easily determined. We will see examples of this later in this lesson.

Rather than relying on statistical theory, we can also use `R` to simulate this situation and show that it is correct. In the code below, we use the `dnorm()` function to calculate the standard normal density and the `dchisq()` function to calculate the $\chi^{2}$ density. We add these as curves on top of the histograms using the `curve()` function

``````r
set.seed(666)
y <- rnorm(n=5000, mean=0, sd=1)

par(mfrow=c(1,2))
par(cex.lab=1.5, cex.axis=1.2, cex=1.2) # plot scaling

hist(y, probability=TRUE, breaks=50, col="skyblue", border="blue")
curve(dnorm(x,mean=0,sd=1), col="red", lwd=5, add=TRUE)

hist(y^2, probability=TRUE, breaks=50, col="skyblue", border="blue")
curve(dchisq(x,df=1), col="red", lwd=5, add=TRUE)
``````
